## Merge Dataset Completo com Latitude e Longitude das cidades da RIDE
Nessa parte será realizado merge entre o arquivo datasetCompleto.csv com o arquivo cod_lat_lon.csv. Duas colunas serão criadas do dataset, relativos a latitude e longitude de cada cidade. O merge será feito utilizando o código do município.
(Verificar se é necessário essa fase)

```
#Lendo o arquivo com todos os meses em um só arquivo csv
df_left = pd.read_csv('/content/drive/My Drive/PIBITI - 2019/DadosTratamento/datasetCompleto.csv',low_memory=True)
```

```
#Verificar a quantidade de linhas e colunas (linhas,colunas)
df_left.groupby('CODIGO_MUNICIPIO_SIAFI').head(1)
```

```
df_left.head()
```

```
#Lendo o arquivo com latitude e longitude de cada cidade da RIDE
df_right = pd.read_csv('/content/drive/My Drive/PIBITI - 2019/DadosTratamento/cod_lat_long.csv')
```

```
df_right.head()
```

```
#Fazendo o merge entre os dataframes df_left e df_right e atribuindo o resultado ao dataframe df_result
df_result = pd.merge(left=df_left, right=df_right, left_on='CODIGO_MUNICIPIO_SIAFI', right_on='CODIGO_MUNICIPIO_SIAFI', how='inner')
```

```
#Verificando a quantidade de linhas e colunas do novo dataframe
df_result.shape
```

```
#Aqui podemos observar as duas colunas LAT E LON referente a cada uma das cidades da RIDE
df_result.groupby('CODIGO_MUNICIPIO_SIAFI').head(1)
```
