## **Visualização região integrada do Distrito Federal**

```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import geopandas as gpd
pd.set_option("display.max_rows", 3000)
```

### Obtendo csv com o valor médio recebido por cidade

```python
df = pd.read_csv('/home/marcos/Desktop/dados/dadosVis/df1.csv')
```

### Lendo shp com as 34 cidades da RIDE

```python
geodf = gpd.read_file('/home/marcos/Desktop/dados/ride_shp/rideTotal2.shp')
geodf.shape
```

```python
#Transformar geodf para um dataframe pandas
df_temp = pd.DataFrame(geodf)
```

### Fazendo o merge entre os dataframes df e df_temp
- Objetivo: criar um dataframe contendo os valores médios por cidade e a coluna geometry para ser possivel plotar um choropleath

```python
df_result =  pd.merge(left=df, right=df_temp,left_on='NOME_MUNICIPIO', right_on='NM_MUNICIP', how='inner')
del df_result['NOME_MUNICIPIO']
df_result.head()
```

```python
#Transformando o df_result para um dataframe do tipo geodataframe
geodataframe = gpd.GeoDataFrame(df_result)
```

### Arquivo GeoJSON

```python
import fiona
import json
```

```python
geodf.to_file('ride.geojson', driver='GeoJSON')
```

```python
#Lendo arquivo geojson
with open('ride.geojson') as file:
    ride = json.load(file)

ride["features"][0]
```

### Plotagem de choropleth

```python
import plotly.express as px
```

```python
fig = px.choropleth(df_result, geojson=ride,color="VALOR_PARCELA",
            locations="NM_MUNICIP",featureidkey="properties.NM_MUNICIP",
            projection="mercator"
            )

fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.shoBw()
```
