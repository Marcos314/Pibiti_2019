# Criando um dataframe com Latitude e Longitude

```python
import pandas as pd
import numpy as np
```

```python
#Lendo o arquivo e criando um dataframe
df = pd.read_csv('/home/marcos/Desktop/dados/clean_data/201302.csv')
#df = df.astype('category')
df.head()
```

```python
del df['MES_REFERENCIA']
del df['CODIGO_MUNICIPIO_SIAFI']
del df['NOME_MUNICIPIO']
del df['NOME_FAVORECIDO']
```

```python
df.head()
```

```python
for index, row in df.iterrows():    
    if row['UF'] == 'GO':
        df.loc[index,'LAT'] = -16.64
        df.loc[index,'LON'] = -49.31
    
    if row['UF'] == 'MG':
        df.loc[index,'LAT'] = -18.10
        df.loc[index,'LON'] = -44.38
    
    if row['UF'] == 'DF':
        df.loc[index,'LAT'] = -15.83
        df.loc[index,'LON'] = -47.86
    
```
