## Juntando arquivos csv

```python
import pandas as pd
pd.set_option("display.max_rows", 2000)
```

```python
#Lendo o arquivo com todos os meses em um s√≥ arquivo csv
df_left = pd.read_csv('/home/marcos/Desktop/dados/clean_data/datasetCompleto.csv',low_memory=True)
```

```python
df_left.head()
```

```python
#Lendo o arquivo com latitude e longitude de cada cidade da RIDE
df_right = pd.read_csv('/home/marcos/Desktop/dados/cod_lat_long.csv')
```

```python
df_right.shape
```

```python
df_right.head()
```

```python
df_result = pd.merge(left=df_left, right=df_right, left_on='CODIGO_MUNICIPIO_SIAFI', right_on='CODIGO_MUNICIPIO_SIAFI', how='inner')
```
